```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('athlete_events.csv')
```

```python
pd.options.display.max_rows = 100
pd.options.display.max_columns = 100
```

```python
df
```

```python
# Which country has the most appearances
df.groupby('Team')['ID'].count().sort_values(ascending=False).head(20).plot(kind='bar')
```

```python
# Which country has the most medals
df.groupby('Team')['Medal'].count().sort_values(ascending=False).head(20).plot(kind='bar')
```

```python
# Medal winners: Gender comparison
df[df['Medal'].notna()].groupby(['Year','Sex'])['ID'].count().reset_index().sort_values('Year', ascending=False).head(10)
```

```python
# America has the most medals. What are the average stats for their winning atheletes in the last 20 years vs average?
df_last20_US = df[(df['Medal'].notna()) & (df['Year'] >= 2004) & (df['Team'] == 'United States')]

df_last20 = df[(df['Medal'].notna()) & (df['Year'] >= 2004) & (df['Team'] != 'United States')]

pd.merge(
    df_last20_US.groupby('Sex')[['Height', 'Weight', 'Age']].mean().round(2),
    df_last20.groupby('Sex')[['Height', 'Weight', 'Age']].mean().round(2),
    on='Sex',
    suffixes=('_US', '_world'))[['Height_US','Height_world','Weight_US','Weight_world','Age_US','Age_world']]
```

```python
# How well has Malaysia performed amongst Olympic medal winners?
df_medal_rank = df[df['Medal'].notna()].drop_duplicates(subset=['Team','Year','Event']
                    ).groupby('Team')['Medal'].count().sort_values(ascending=False).reset_index()

df_medal_rank[df_medal_rank['Team'] == 'Malaysia']
df_medal_rank['Ranking'] = df_medal_rank['Medal'].rank(method='dense', ascending=False)

pd.concat(
    [df_medal_rank.head(1),
    df_medal_rank[df_medal_rank['Team'] == 'Malaysia'],
    df_medal_rank.tail(1)])

# Malaysia was Rank 72 / 80 amongst medal winners!
```

```python
# What events are Malaysia's wins in?

df[(df['Medal'].notna()) & (df['Team'] == 'Malaysia')].drop_duplicates(subset=['Year', 'Event'])['Sport'].value_counts(
    ).plot(kind='pie', title="Malaysia's winning categories")
```
